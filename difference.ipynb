```python
import torch

# 
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)

#  y1
y1 = x ** 2
y1 = torch.linalg.norm(torch.ones_like(y1)-y1)
#  y1  x 
grad_y1 = torch.autograd.grad(outputs=y1, inputs=x, retain_graph=True)
print("Gradient of y1:", grad_y1[0]) 

#  x 
# x.grad = None  #  x.grad.zero_()

#  y2
y2 = torch.linalg.norm(torch.zeros_like(y1)-y1)
#  y2  x 
grad_y2 = torch.autograd.grad(outputs=y2, inputs=x)
print("Gradient of y2:", grad_y2[0])  # : tensor([3., 12., 27.])
```
